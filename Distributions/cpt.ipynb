Импортируем нужные нам библиотеки.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
import statistics
%matplotlib inline
```

Я выбираю гамма распределение, поэтому ниже создадим выборку в 1000 элементов из гамма распределения. 

```python
gamma_rv = sts.gamma(1.99)
sample = gamma_rv.rvs(1000)
```

Нарисуем гистограмму выборки, а поверх нее нарисуем график функции плотности гамма - распределения

```python
plt.hist(sample, bins = 20, density = True, histtype = 'bar', ec = 'black')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')

x = np.linspace(-10, 10, 1000)
pdf = gamma_rv.pdf(x)
plt.plot(x, pdf, label='theoretical PDF')
plt.legend(loc='upper right')
```

Теперь создадим 1000 выборок по 5, 10 и 50 элементов из гамма - распределения и запишем их средне - арифметические

```python
averages_of_fives = []
averages_of_tens = []
averages_of_fifties =[] #списки, где хранятся средние значения выборок
for i in range(1000):
    sample = gamma_rv.rvs(5)
    averages_of_fives.append(np.mean(sample))
for i in range(1000):
    sample = gamma_rv.rvs(10)
    averages_of_tens.append(np.mean(sample))
for i in range(1000):
    sample = gamma_rv.rvs(50)
    averages_of_fifties.append(np.mean(sample))
        
```

У нашего гамма распределения следующие параметры: a = 1 b = 1,99. Значит математическое ожидание = b/a = 1,99, дисперсия = b/(a^2) = 1,99. Cогласно ЦПТ распределения средних близки к следующим нормальным распределениям: для 5 - N(1.99, 1.99/5),
для 10 - N(1.99, 1.99/10), для 50 - N(1.99, 1.99/50). Нарисуем гистограммы выборок и поверх них нарисуем плотности соответствующих гамма распределений.

```python
norm_rv1 = sts.norm(loc = 1.99, scale = 1.99 /math.sqrt(5))
norm_rv2 = sts.norm(loc = 1.99, scale = 1.99 / math.sqrt(10)) # нормальные распределения 
norm_rv3 = sts.norm(loc = 1.99, scale = 1.99 / math.sqrt(50)) 
```

```python
plt.hist(averages_of_fives, bins = 20, density = True, histtype = 'bar', ec = 'black')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')

x = np.linspace(-10, 10, 1000)
pdf = norm_rv1.pdf(x)
plt.plot(x, pdf, label='PDF of normal RV')
plt.legend(loc='upper right')
```

```python
plt.hist(averages_of_tens, bins = 20, density = True, histtype = 'bar', ec = 'black')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')

x = np.linspace(-20, 20, 1000)
pdf = norm_rv2.pdf(x)
plt.plot(x, pdf, label='PDF of normal RV')
plt.legend(loc='upper right')
```

```python
plt.hist(averages_of_fifties, bins = 20, density = True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')

x = np.linspace(0, 10, 1000)
pdf = norm_rv3.pdf(x)
plt.plot(x, pdf, label='PDF of normal RV')
plt.legend(loc='upper right')
```
