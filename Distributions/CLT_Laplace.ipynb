```python
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
from math import exp, sqrt

%matplotlib inline
```

### The probability density function for `laplace` (https://en.wikipedia.org/wiki/Laplace_distribution) is:

$$ f(x, | \mu, b) = \frac{1}{2b}e^{-\frac{|x - \mu|}{b}}$$

### mean

$$ \mu = \mu$$


### variance

 $$ D = 2b^2$$

```python
# Задаем функцию теоретической плотности распределения случайной величины
# Внутри функции задаем расчет дисперсии и мат. ожидания 
# в зависимости от задаваемых параметров функции

def theor_pdf(x, m, b):
    variance = 2*b**2
    mu = m
    return (np.array([1/(2*b)*exp(-1*abs((x_i-m)/b)) for x_i in x]), variance, mu)

m = 0
b = 1
x = np.linspace(0, 3, 1000)

variance = theor_pdf(x, m, b)[1]
mu = theor_pdf(x, m, b)[2]

print('Variance: ', variance)
print('mu: ', mu)
```

```python
# Гистограмма выборки объёма 1000 распределения Лапласа (scipy.stats.laplace)
laplace = sts.laplace.rvs(size=1000)

fig, ax = plt.subplots(1, 1)

ax.hist(laplace, density=True, label='rvs')

# Теоретическая плотность распределения
x = np.linspace(-8, 8, 1000)
ax.plot(x, theor_pdf(x, m, b)[0], 'r-', lw=5, alpha=0.6, label='theor_pdf')
plt.title('Распределение Лапласа (scipy.stats.laplace)')
plt.xlabel('X', color='black')
plt.ylabel('Result', color='black')

plt.legend()
plt.tight_layout()
```

```python
# Гистограмма распределений выборочных средних 1000 выборок объёма 5
n = 5
means_list = []
for i in range(0, 1000):
    means_list.append(np.mean(sts.laplace.rvs(size=n)))

fig, ax = plt.subplots(1, 1)

ax.hist(np.array(means_list), density=True, histtype='stepfilled', 
        color='b', stacked=False, label='means')

# Плотность нормального распределения Лапласа 
# с расчетными параметрами функции (mu, variance)
x = np.linspace(-3, 3, 1000) 
rail = sts.laplace(mu, sqrt(variance/n))
r = rail.pdf(x)
ax.plot(x, r, 'r', label='pdf')
plt.title('Размер выборки: {}'.format(n))
plt.xlabel('X', color='black')
plt.ylabel('Result', color='black')

plt.legend()
plt.tight_layout()
```
