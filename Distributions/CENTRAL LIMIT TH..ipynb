```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as pyplot
```

Выполним задание рассматривая распределение Стьюдента.
Его матожидание равно 0 при n>1, а его дисперсия равна n/(n-2) при n>2

```python
n = 1000 #размер нашей выборки
k = 5 # параметр для функций в stats
x = stats.t(k)
set = x.rvs(n) # так мы генерируем набор для работы
#print(set)
```

```python
y = np.linspace(-30, 30, 200)
pyplot.hist(set, density=True, bins = 22, histtype= 'bar', ec='black') #рисуем гистограмму на наборе set
pyplot.plot(y, x.pdf(y))#рисуем probability density function
pyplot.title('Гистограмма выборки set (blue bars) и теоретическая плотность распределения(желтая линия)')
pyplot.xlabel('$y$')
pyplot.ylabel('$pdf$')
```

Попробуем выборки меньшего фиксированного объёма.

Объём выборки n = 5 :
    

```python
n = 5
k = 5
x = stats.t(k)
set = x.rvs(n)
print(set)
```

Создаем случайную выборку выборочных средних:
    

```python
#здесь сделаем список со средними выборочных и заполним его
arr_mean = [] #инициализация
for i in range(1000):
    arr_mean.append(sum(x.rvs(n))/n) # заносим значение в список

arr_mean = np.array(arr_mean)
```

```python
pyplot.hist(arr_mean, density=True, bins = 20, histtype='bar', ec='black') # строим гистограмму по выборке выборочных средних
pyplot.title('Гистограмма распределения выборочных средних при $n=5$ и \n плотность нормального распределения с соответствующими параметрами (по ЦПТ)')

norm = stats.norm(0, np.sqrt(1/(n-2))) # создаем случайную нормально распределенную величину
y = np.linspace(-7, 7, 100)
pyplot.plot(y, norm.pdf(y))
pyplot.xlabel('y')
```
